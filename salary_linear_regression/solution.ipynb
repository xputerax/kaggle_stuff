```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
SEED = 42

X = df['YearsExperience'].values
Y = df['Salary'].values

# https://stackoverflow.com/questions/51150153/valueerror-expected-2d-array-got-1d-array-instead
X = X.reshape(-1, 1)
Y = Y.reshape(-1, 1)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=SEED, test_size=.5, shuffle=True)
```

```python
print(X.shape)
print(Y.shape)
```

```python
model = LinearRegression()
model.fit(X=X_train, y=Y_train)
```

```python
model.score(X=X_test, y=Y_test)
```

```python
Y_pred = model.predict(X_test)
```

```python
plt.scatter(X_train, Y_train, color='g') # plot the dots
plt.plot(X_test, Y_pred, color='k') # draw the line
plt.show()
```
